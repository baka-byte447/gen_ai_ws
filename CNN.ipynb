## CNN - Convulution Neural Network

```python
import tensorflow as tf

import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.DataFrame({
    "soil_moisture" : [0.10, 0.15, 0.20, 0.25, 0.30, 0.35, 0.40, 0.50, 0.60, 0.18, 0.45, 0.80],
    "temperature_c" : [34, 30, 26, 22, 28, 30, 19, 22, 25, 33, 12, 21],
    "sunlight_hours" : [9, 8, 7, 4, 8, 10, 12, 5, 9, 11, 3, 9],
    "needs_water" : [1, 1, 1, 1, 0, 0, 0, 0, 1, 0, 1, 0],
})
df
```

```python
df.columns
```

```python
X = df[['soil_moisture', 'temperature_c', 'sunlight_hours']]
y = df['needs_water']
```

```python
X_min = X.min()
X_max = X.max()
```

```python
# mean median mode

X_scaled = (X-X_min) / (X_max-X_min + 1e-8)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=42,stratify=y)
```

```python
from tensorflow.keras import layers
model = keras.Sequential(
    [
        layers.Input(shape = (X_train.shape[1], )),
        layers.Dense(8,activation='relu'),
        layers.Dense(1,activation='sigmoid')

    ]

)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python

history_full = model.fit(X_train.values, y_train.values.reshape(-1,1), epochs=100, batch_size=50, verbose=1)

```

```python
history = model.fit(
    X_train.values,y_train.values.reshape(-1,1),
    validation_data = (X_test.values, y_test.values.reshape(-1,1)),
    epochs = 100,
    batch_size = 1,
    verbose=1
)
```
